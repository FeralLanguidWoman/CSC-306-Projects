### 2. Making the reverse complement
Since genes can also occur on the other strand of DNA (also in the 5'->3' direction), we need to look for genes in the reverse complement too. In the block below, write a function (or you can copy-and-paste from your homework) that takes as input a string of DNA and returns a string that is the reverse complement. Call this function with `dna_string` as your input and store the output in a variable called `r_dna_string`. Again, check its length, which should also be `6387`.

```python
# Pseudocode:

# start by defining a fuction to give the reverse complement of a given dna sequence
# the parameter will be "dna_string"
  # within the function:
    # create an empty string called "r_dna_string"
    # make a for loop "for x in dna":
      # within the for loop:
        # if x equals A, add T to the reverse compliment
        # if x equals T, add A to the reverse compliment
        # if x equals C, add G to the reverse compliment
        #if x equals G, add C to the reverse compliment
    # return "r_dna_string"
# call the function
```

```python
# Hint: you can reuse your code from a previous HW!
```

```python


def reverse_complement(dna_string): # creates a function to give the reverse compliment of a dna_string
  '''creates a reverse compliment of an inputed string of dna'''
  r_dna_string = '' # creates an empty string for the reverse compliment
  dna_string = dna_string[::-1]
  for x in dna_string:  # creates a for loop to iterate over the given dna sequence
    if x == 'A':  # if x equals A, add T to the reverse compliment
      r_dna_string += 'T'
    if x == 'T':  # if x equals T, add A to the reverse compliment
      r_dna_string += 'A'
    if x == 'C':  # if x equals C, add G to the reverse compliment
      r_dna_string += 'G'
    if x == 'G':  # if x equals G, add C to the reverse compliment
      r_dna_string += 'C'
  return r_dna_string # returns the complited string for the reverse compliment

reverse_complement(dna_string) # call the function with the string you want the reverse compliment of
```

```python
# Pseudocode:

# define a fuction "orf_finder" with the parameter "dna_sequence":
  # create an empty list called "orf_list"
  # create a boolean varible: "in_orf" equals false
  # create an empty string "current_orfs" to hold each orf string
  # create a for loop: for i in a range of zero to the length of the dna_sequence:
    # create a variable "codons" equals dna_sequence[i:i +3]
    # if the codon is "ATG":
      # then in_orf equals true
    # if in_ orf (if in_orf = true):
      # add the codons to the "current_orf" string
    # if the codon is "TAA", "TAG", "TGA":
      # if in_ orf (if in_orf = true):
      # append the "orf_list" to add the "current_orf" to it
      # reset the "current_orf" string back to empty
      # in_orf equals false
  #return the orf_list

# call the function
```

```python
def orf_finder(dna_sequence):  # defines function "orf_finder"
  '''finds orfs in the inputed string of dna'''
  orf_list = []  # creates an empty list to hold the orf strings
  in_orf = False    # starts us outside of the orf
  current_orf = ''  # create an empty string to make each orf string

  for i in range(0, len(dna_sequence), 3):  # creates a for loop to iterate over the dna sequence
    codons = dna_sequence[i:i + 3] # splits the dna sequence into codons
    if codons in 'ATG':  # defines the the start of the orf using the start codon "ATG"
      in_orf = True # if a codon is "ATG" then its "True" we're in an orf
    if in_orf:   # if we're in an orf, then:
      current_orf += codons # add codons to the current_orf string
    if codons in ['TAA','TAG','TGA']:  # when iterating through the codons if a stop codon, "TAA", "TAG", or "TGA" is reached:
      if in_orf:    # if the list is still in an orf:
        orf_list.append(current_orf)  # append the orf list to add what is contained in the current orf string
        current_orf = '' # reset the current orf string back to empty
        in_orf = False  # reset out of the orf

  return orf_list # once the dna sequence iterated,


orf_finder(dna_string) # calls the orf_finder function

```

### 6. Outputting your list of genes in a fasta file
This one is easy - we've provided a function that takes as input a list of genes and an output filename (your output filename should be `X73525_genes.fasta`) and write the gene sequences to a fasta file you can use on the BLAST website.

```python
#Pseudocode:

# I'm assuming that I don't need to write pseudocode for this fuction as I did not need to write it.

#call fasta_writer function with inputs amino_acids, 'X73525_genes.fasta'
```

```python
def fasta_writer(gene_list, output_filename):
    '''
    given a list of genes and an output_filename, output those genes to a fasta file
    '''
    with open(output_filename, 'w') as f:
        gene_counter = 1
        for gene in gene_list:
            f.write('>gene_'+str(gene_counter)+'\n')
            for i in range(len(gene)//80+1): # writes 80 characters per line
                f.write(gene[i*80:i*80+80] + '\n')
            gene_counter += 1

```

```python
def gene_finder(file, new_filename): # defines the function "gene_finder" and sets the parameter as "file"
  '''puts a series of functions together to create a function that reads a file of dna,
  puts that dna into a string, finds the reverse compliment of the dna and puts that into a string,
  finds the orfs in the dna string and the reverse compliment string, combines all of the orf strings
  into one list, finds the possible genes in the orf list,  then translates the codons in those genes
  into amino acids, and then inputs those amino acids into a fasta file
  '''

  '''tldr: takes a sequence of dna from a file, finds the genes, translates the genes into amino acids,
  and puts that information into a new file
  '''

  dna_string = read_fasta(file)  # sets the variable "dna_string" as equal to the function read_fasta
  r_dna_string = reverse_complement(dna_string) # sets the variable "r_dna_string" as equal to the function reverse_compliment
  orfs1 = orf_finder(dna_string[0:]) # finds orfs for reading frame 1 assigns it to variable "orfs1"
  orfs2 = orf_finder(dna_string[1:]) # finds orfs for reading frame 2 assigns it to variable "orfs2"
  orfs3 = orf_finder(dna_string[2:]) # finds orfs for reading frame 3 assigns it to variable "orfs3"
  r_orfs1 = orf_finder(r_dna_string[0:]) # finds orfs for reading frame 4 assigns it to variable "r_orfs1"
  r_orfs2 = orf_finder(r_dna_string[1:]) # finds orfs for reading frame 5 assigns it to variable "r_orfs2"
  r_orfs3 = orf_finder(r_dna_string[2:]) # finds orfs for reading frame 6 assigns it to variable "r_orfs3"
  all_orfs = orfs1 + orfs2 + orfs3 + r_orfs1 + r_orfs2 + r_orfs3 # combines the orfs from all reading frames into one list
  amino_acids = translate(all_orfs) # assigns variable "amino_acids" to the function translate
  new_file = fasta_writer(amino_acids, new_filename) # sets the variable "new_file" to the function fasta_writer
                                            # it would need a name different from the name used for the data we used in this project
  return new_file # returns a fasta file with the amino acids translated from the genes found in the the original dna sequence
```
